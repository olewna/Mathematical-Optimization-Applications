# LAB 13

Input $p = (...)$ length n

$m = (m^0, m^1, m^2, m^3...)$

$p(x) = p'm = \Sigma_{i=0}^{n} p_i * x^i$

If $p(x) = \Sigma_{i=1}^{r} s_i^2 => p(x) = m^TAm, A>=0$

find $A \in \R^{n_xn}$

```julia
using JuMP
using SDPA
using SCIP
using LinearAlgebra
using SCS
using DynamicPolynomials
```

```julia
@polyvar x
p = [1,2,1,0,1]
d = 4
m = monomials(x, 0:d)
px = p'*m
```

```julia
model = Model(SCS.Optimizer)
```

```julia
deg = length(p) - 1
d = div(deg, 2)

@variable(model, A[1:d+1, 1:d+1])
```

```julia
@objective(model, Min, 0)
```

```julia
@constraint(model, A .>= 0)
```

```julia
for k = 0:deg
    expr = 0.0
    for i = 0:d
        for j = 0:d
            if i+j == k
                expr += A[i+1, j+1]
            end
        end
    end
    @constraint(model, expr == p[k+1])
end
```

```julia
set_silent(model) #  shhh
optimize!(model)
```
